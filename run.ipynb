```python
import subprocess
from tqdm import tqdm

import warnings
import os

# 경고 메시지 무시 설정
warnings.filterwarnings("ignore", category=UserWarning)

# 환경 변수 설정
os.environ['PYTHONPATH'] = os.getcwd()

# inference.py 실행
process = subprocess.Popen([
    'python', 'inference.py',
    '--dataset', 'vitonhd',
    '--data_root_path', r'C:\Users\epdgn\Downloads\catvton\VITON-HD',
    '--output_dir', r'C:\Users\epdgn\Downloads\catvton\output',
    '--dataloader_num_workers', '2',
    '--batch_size', '1',
    '--seed', '555'
], stdout=subprocess.PIPE, stderr=subprocess.STDOUT, text=True, universal_newlines=True, encoding='utf-8')

# 실시간 로그 출력
for line in iter(process.stdout.readline, ""):
    if line:
        tqdm.write(line.strip())
    else:
        break
process.stdout.close()
process.wait()

print("Inference process completed.")
```
